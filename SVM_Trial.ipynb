```python
#implementing the necessary packages
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from PrimalSVM import PrimalSVM
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
from matplotlib import pyplot as plt
import numpy as np
from mlxtend.plotting import plot_decision_regions

# logistic regression tends to involve a lot of log(0) and things that wash out in the end. 
np.seterr(all='ignore') 
np.random.seed(12345)
```

```python
def pad(X):
    return np.append(X, np.ones((X.shape[0],1)), axis=1)
```

# Testing for Linear Data

```python
np.random.seed(12345)
X, y = make_blobs(n_samples = 200, n_features = 2, centers = [(-1,-1),(1,1)])
PS = PrimalSVM(0.1)
```

```python
PS.fit(X, y, 0.1, 1000, 50)
PS.score_history[-1]
```

```python
plot_decision_regions(X, y, clf = PS)
t = title = plt.gca().set(title = f"Accuracy = {PS.score(X, y)}",
                      xlabel = "Feature 1", 
                      ylabel = "Feature 2")
```

# Testing Logistic Regression

```python
np.random.seed(12345)
X, y = make_blobs(n_samples = 200, n_features = 2, centers = [(-1,-1),(1,1)])
LR = LogisticRegression()
```

```python
LR.fit(X, y)
LR.score(X, y)
```

```python
plot_decision_regions(X, y, clf = LR)
t = title = plt.gca().set(title = f"Accuracy = {LR.score(X, y)}",
                      xlabel = "Feature 1", 
                      ylabel = "Feature 2")
```

# Non-Linear Data

```python
np.random.seed(12345)
X, y = make_moons(200, shuffle = True, noise = 0.2)
SGD = SGDClassifier(loss='hinge', penalty='l2')
```

```python
SGD.fit(X, y)
SGD.score(X, y)
```

```python
plot_decision_regions(X, y, clf = SGD)
t = title = plt.gca().set(title = f"Accuracy = {SGD.score(X, y)}",
                      xlabel = "Feature 1", 
                      ylabel = "Feature 2")
```
